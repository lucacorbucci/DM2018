```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn import preprocessing


df = pd.read_csv("credit_default_train.csv")
```

```python
correlation = df.corr()
```

```python

sns.set(style="white")

# Generate a mask for the upper triangle
mask = np.zeros_like(correlation, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
fig = plt.figure(figsize=(12,12))
fig_dims = (3,2)

plt.subplot2grid(fig_dims, (0, 0))
df['education'].value_counts().plot(kind='bar', 
                                   title='Education')
plt.xticks(rotation=0)

plt.subplot2grid(fig_dims, (0,1))
df['sex'].value_counts().plot(kind='bar', 
                                   title='Sex')
plt.xticks(rotation=0)

plt.subplot2grid(fig_dims, (1,0))
df['status'].value_counts().plot(kind='bar', 
                                   title='Status')
plt.xticks(rotation=0)
plt.show()
```

```python
df['age'].hist(bins=15)
plt.title("Istogramma eta'")
```

```python
df['age'].value_counts().hist()
```

```python
# Grafico in cui vediamo le persone che hanno o non hanno credit default rispetto all'etÃ .
# In particolare in verde ci sono le persone che hanno avuto un credit default e in rosso quelle che 
# invece non hanno il credit default

plt.scatter(df[df['credit_default'] == 'yes']['age'].unique(), 
            df[df['credit_default'] == 'yes']['age'].value_counts(), color='g', marker='*')

plt.scatter(df[df['credit_default'] == 'no']['age'].unique(), 
            df[df['credit_default'] == 'no']['age'].value_counts(), color='r', marker='+')

plt.show()
```

```python
# Pclass
pclass_xt = pd.crosstab(df['age'], df['credit_default'])


fig, axes = plt.subplots(figsize=(15,8))

pclass_xt.plot(kind='bar', stacked=True,ax=axes, 
                   title="Situazione del credit default rispetto all'eta'")
plt.xlabel('Eta')
plt.ylabel('Frequenza credit_default')
plt.xticks(rotation=0)
plt.show()
```
